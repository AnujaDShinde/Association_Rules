```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
import numpy as np
import matplotlib.pyplot as plt
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
books = pd.read_csv('book.csv')
books
```

```python
# No preprocessing is required
```

```python
# Apriori Algorithm
```

```python
# Value of support helps us to identify the rules worth considering for further analysis.
# This measure defines the likeliness of occurrence of consequent on the cart given that the cart already has the antecedents.
# Lift is the measure that will help store managers to decide product placements on aisle.
```

```python
# Association rule with support = 5 % , confidence = 75 %
```

```python
# support = 5 %

frequent_itemsets = apriori(books, min_support = 0.1, use_colnames = True)
frequent_itemsets
```

```python
# confidence = 75 %

rules = association_rules(frequent_itemsets, metric = 'lift')
rules
```

```python
rules.sort_values('lift', ascending = False)
```

```python
# Lift Ratio > 1 is a good influential rule in selecting the associated transactions

rules[rules.lift > 1]
```

```python
# Visualization of obtained rule

plt.scatter(rules['support'], rules['confidence'])
plt.xlabel('support')
plt.ylabel('confidence') 
plt.show()
```

```python
# From above scatter plot => linear positive association
```
